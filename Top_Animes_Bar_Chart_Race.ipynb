```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import bar_chart_race as bcr
```

```python
pytrends = TrendReq(hl='en-US')
keywords = ['Fullmetal Alchemist: Brotherhood','Gintama','Attack on Titan','Death Note','Demon Slayer']
pytrends.build_payload(keywords,timeframe ='today 5-y')
```

```python
data = pytrends.interest_over_time()
data = data.drop('isPartial', axis=1)

data.head(10)
```

```python
type(data)
```

```python
plt.plot(data)
plt.title('Top 5 Animes on Google Search')
plt.xlabel('Year')
plt.ylabel('Weekly Searches')
plt.legend(keywords,loc='right')
plt.show()
```

```python
bcr.bar_chart_race(df = data, title = "Top 5 Animes on Google Search")
```
